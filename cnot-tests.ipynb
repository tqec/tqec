```python
from tqec.gallery import cnot
from tqec.interop.pyzx.plot import plot_positioned_zx_graph, draw_correlation_surface_on
from tqec.interop import PositionedZX
from tqec import Position3D

from pyzx.graph.graph_s import GraphS
from pyzx import VertexType
from tqec.interop.pyzx.synthesis import block_synthesis as block_synthesis
```

```python
graph = cnot()
# graph.view_as_html()
zx_graph = graph.to_zx_graph()
zx_graph.draw()
correlation_surfaces = graph.find_correlation_surfaces()
fig, ax = zx_graph.draw()
draw_correlation_surface_on(correlation_surfaces[1], zx_graph, ax)
```

```python
g_zx = GraphS()

g_zx.add_vertices(10)
g_zx.set_type(2, VertexType.Z)
g_zx.set_type(3, VertexType.X)
g_zx.set_type(4, VertexType.Z)
g_zx.set_type(5, VertexType.X)
g_zx.set_type(6, VertexType.X)
g_zx.set_type(7, VertexType.X)
g_zx.add_edges([(0, 2), (1, 3), (2, 4), (4, 5), (5, 6), (3, 6), (2, 7), (7, 8), (6, 9)])
g_zx.set_inputs(
    (
        0,
        1,
    )
)
g_zx.set_outputs(
    (
        8,
        9,
    )
)

positions = {
    0: Position3D(0, 0, 0),
    1: Position3D(1, 1, 0),
    2: Position3D(0, 0, 1),
    3: Position3D(1, 1, 1),
    4: Position3D(0, 1, 1),
    5: Position3D(0, 1, 2),
    6: Position3D(1, 1, 2),
    7: Position3D(0, 0, 2),
    8: Position3D(0, 0, 3),
    9: Position3D(1, 1, 3),
}

g = block_synthesis(g_zx, positions=positions)
zx_graph = PositionedZX(g_zx, positions)
plot_positioned_zx_graph(zx_graph)
```

```python
from tqec.interop.pyzx.plot import plot_positioned_zx_graph, draw_correlation_surface_on
from tqec.interop import PositionedZX

zx_graph = PositionedZX(g_zx, positions)
correlation_surfaces = g.find_correlation_surfaces()
fig, ax = plot_positioned_zx_graph(zx_graph)
draw_correlation_surface_on(correlation_surfaces[0], zx_graph, ax)
```

```python
# graph is the cnot BlockGraph
# g is the hand CNOT GraphS synthesized as a BlockGraph

print(graph.__eq__(g))
```
